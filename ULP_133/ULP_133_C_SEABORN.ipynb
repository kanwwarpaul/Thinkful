```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("flavors_of_cacao.csv")
x = df["Cocoa\nPercent"].str[0:2]
df["Cocoa\nPercent"] = x.astype(int)
```

```python
plt.figure(figsize=(12,6))
sns.stripplot(x="Review\nDate", y="Rating", data=df)
```

```python
plt.figure(figsize=(12,6))
sns.swarmplot(x="Review\nDate", y="Rating", data=df)
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x="Review\nDate", y="Rating", data=df)
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x="Review\nDate", y="Rating", data=df)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Review\nDate", y="Rating", data=df)
```

```python
plt.figure(figsize=(12,6))
sns.pointplot(df["Cocoa\nPercent"],df["Rating"])
plt.show()
```
