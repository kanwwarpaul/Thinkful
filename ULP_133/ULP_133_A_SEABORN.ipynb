```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("flavors_of_cacao.csv")
df.head()
```

```python
x = df["Cocoa\nPercent"].str[0:2]
df["Cocoa\nPercent"] = x.astype(int)
```

```python
df.describe()
```

```python
plt.figure(figsize=(12,6))
sns.distplot(df["Cocoa\nPercent"])
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.distplot(df["Cocoa\nPercent"],kde=False)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.distplot(df["Cocoa\nPercent"],rug=True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.distplot(df["Cocoa\nPercent"],kde=False,rug=True)
plt.show()
### A histogram represents the distribution of data by forming bins along the range of the data 
### and then drawing bars to show the number of observations that fall in each bin.
### To illustrate this, let’s remove the density curve and add a rug plot, 
### which draws a small vertical tick at each observation.
```

```python
plt.figure(figsize=(12,6))
sns.kdeplot(df["Cocoa\nPercent"])
sns.rugplot(df["Cocoa\nPercent"])
plt.show()
### it can be a useful tool for plotting the shape of a distribution.
### Like the histogram, the KDE plots encodes the density of observations on one axis with height along the other axis.
```

```python
plt.figure(figsize=(12,6))
sns.kdeplot(df["Cocoa\nPercent"],shade=True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.kdeplot(df["Cocoa\nPercent"])
plt.show()
```
