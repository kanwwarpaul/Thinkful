```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("flavors_of_cacao.csv")
x = df["Cocoa\nPercent"].str[0:2]
df["Cocoa\nPercent"] = x.astype(int)
```

```python
plt.figure(figsize=(12,6))
sns.jointplot(df["Cocoa\nPercent"],df["Rating"])
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.jointplot(df["Cocoa\nPercent"],df["Rating"],kind="reg")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.jointplot(df["Cocoa\nPercent"],df["Rating"],kind="hex")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.jointplot(df["Cocoa\nPercent"],df["Rating"],kind="kde")
plt.show()
```

```python
sns.pairplot(df)
```

```python
g = sns.PairGrid(df)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot, cmap="Blues_d", n_levels=6);
```

```python
plt.figure(figsize=(12,6))
sns.regplot(df["Cocoa\nPercent"],df["Rating"])
plt.show()
```
